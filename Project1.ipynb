```python
# 3.Predominant Age Group Among Employees
```

```python
bins = [20, 30, 40, 50, 60]  
labels = ['20-30', '30-40', '40-50', '50-60']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
```

```python
age_group_counts = df['Age Group'].value_counts()
```

```python
age_group_counts
```

```python
plt.figure(figsize=(7,5))
age_group_counts.plot(kind='bar', color='green')
plt.title("Predominant Age Group")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.show()
```

The employees in the dataset are distributed across four distinct age groups:
The 20-30 age group represents the vast majority of employees, with 334 employees.
The 30-40 age group comes next, with 119 employees .
The 40-50 and 50-60 age groups have significantly fewer employees, with only 3 employees in the 40-50 group and no employees in the 50-60 group.

```python
# 4.Highest Salary Expenditure by Team and Position
```

```python
team_salary = df.groupby('Team')['Salary'].sum().sort_values(ascending=False)
```

```python
team_salary
```

```python
position_salary = df.groupby('Position')['Salary'].sum().sort_values(ascending=False)
```

```python
position_salary
```

```python
team_salary_percentage = team_salary / team_salary.sum() * 100 
```

```python
team_salary_percentage
```

```python
plt.figure(figsize=(7,7))
team_salary_percentage.plot(kind='pie', autopct='%1.1f%%')
plt.title("Salary Distribution per Team")
plt.ylabel('')  
plt.show()
```

We first group the salary data by teams and sum up the total salary for each one. This helps us understand which department or unit has the highest salary expenditure 
Key Insight:

The team with the highest total salary expenditure is is Cleveland Cavaliers with total salary 109824875.0, followed by Los Angeles Clippers and Oklahoma City Thunder.

```python
position_salary_percentage = position_salary / position_salary.sum() * 100
```

```python
position_salary_percentage
```

```python
plt.figure(figsize=(7,7))
position_salary_percentage.plot(kind='pie', autopct='%1.1f%%')
plt.title("Salary Distribution per Position")
plt.ylabel('') 
plt.show()
```

We analyze salary distribution based on  positions. By grouping salaries by titles and summing them up, we determine which role incurs the highest salary costs.

Key Insight:
Centers (C) receive the highest salary expenditure (21.28%)
Point Guards (PG) are the second highest-paid position (20.90%)
Shooting Guards (SG) have the lowest total salary allocation (18.50%)

```python
# 5.Correlation Between Age and Salary
```

```python
correlation_matrix = df[['Age', 'Salary']].corr()
```

```python
correlation_matrix 
```

```python
plt.figure(figsize=(6,5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix: Age vs Salary")
plt.show()
```

A correlation matrix provides a numerical representation of relationships between variables. Since we are focusing on the correlation between Age and Salary, we extract relevant insights from the matrix.
The correlation coefficient between Age and Salary is 0.209, indicating a weak positive correlation.
